```python
import pandas as pd
data = pd.read_csv('BLE_RSSI_dataset/iBeacon_RSSI_Labeled.csv')
data.head()
```

```python
from sklearn.cluster import DBSCAN

x = data[['b3001','b3002','b3003','b3004','b3005','b3006','b3007','b3008','b3009','b3010','b3011','b3012','b3013']]
db = DBSCAN(eps=0.8,min_samples=3).fit(x)

labels=db.labels_
data['cluster_db'] = labels
data.sort_values('cluster_db')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
from sklearn import datasets, manifold
```

```python
x = data[['b3001','b3002','b3003','b3004','b3005','b3006','b3007','b3008','b3009','b3010','b3011','b3012','b3013']]
y = data[['cluster_db']]
# Scale and visualize the embedding vectors                            
def plot_embedding(X, title=None):     
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)                           

    plt.figure()              
    ax = plt.subplot(111)      
    for i in range(X.shape[0]):                          
        plt.text(X[i, 0], X[i, 1], str(y['cluster_db'][i]),   
                 color=plt.cm.Set1((y['cluster_db'][i]) / 86.),         
                 fontdict={'weight': 'bold', 'size': 9})
    plt.xticks([]), plt.yticks([])
    if title is not None:
        plt.title(title)
```

```python

# t-SNE embedding of the digits dataset
print("Computing t-SNE embedding")           
tsne = manifold.TSNE(n_components=2, init='pca', random_state=0)
t0 = time()
X_tsne = tsne.fit_transform(x)

plot_embedding(X_tsne)                           

plt.show() 
```

```python
x_min, x_max = np.min(X_tsne, 0), np.max(X_tsne, 0)
X = (X_tsne - x_min) / (x_max - x_min) 
plt.figure(figsize=(8,8))
plt.scatter(X[:,0], X[:,1],s=5,c=tuple(y['cluster_db']),cmap=plt.cm.Spectral)
plt.title('basic scatter plot ')
plt.xlabel('variables x')
plt.ylabel('variables y')

plt.legend(loc='upper right')

plt.show()
```
